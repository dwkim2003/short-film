```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return """
<!DOCTYPE html>
<html lang="ko">
<head>
<meta charset="UTF-8">
<title>Short Film</title>

<style>

*{
    margin:0;
    padding:0;
    box-sizing:border-box;
}

body{
    background:#000;
    color:#fff;
    font-family:"Noto Sans KR",sans-serif;
    display:flex;
    justify-content:center;
    align-items:center;
    height:100vh;
    overflow:hidden;
}

#container{
    text-align:center;
    width:90%;
}

h1{
    font-size:52px;
    font-weight:300;
    line-height:1.6;
    opacity:0;
    animation:fadeIn 2s forwards;
}

button{

    width:150px;
    height:55px;

    margin:30px 12px;

    font-size:20px;

    background:white;
    color:black;

    border:none;
    border-radius:8px;

    cursor:pointer;

    transition:.3s;
}

button:hover{

    transform:scale(1.08);

}

.fadeOut{

    animation:fadeOut 1.5s forwards;

}

.fadeIn{

    animation:fadeIn 1.5s forwards;

}

@keyframes fadeIn{

from{
opacity:0;
}

to{
opacity:1;
}

}

@keyframes fadeOut{

from{
opacity:1;
}

to{
opacity:0;
}

}

</style>

</head>

<body>

<div id="container">

<h1 id="text">

죽음을 <br>
선택하시겠습니까?

</h1>

<div id="buttons">

<button onclick="startMovie()">YES</button>

<button onclick="startMovie()">NO</button>

</div>

</div>

<script>

const text=document.getElementById("text");
const buttons=document.getElementById("buttons");

function changeMessage(message){

    text.className="fadeOut";

    setTimeout(()=>{

        text.innerHTML=message;

        text.className="fadeIn";

    },1500);

}

function startMovie(){

    buttons.style.display="none";

    setTimeout(()=>{
        changeMessage("오늘 당신의 이야기를<br>끝까지 들어준 사람은,<br><br><strong>여기 있습니다.</strong>");
    },500);

    setTimeout(()=>{
        changeMessage("고생 많았습니다.");
    },5000);


    setTimeout(()=>{
        changeMessage("3");
    },9000);

    setTimeout(()=>{
        changeMessage("2");
    },13000);

    setTimeout(()=>{
        changeMessage("1");
    },15000);


    setTimeout(()=>{

        document.body.style.background="#000";
        document.getElementById("container").innerHTML="";

    },19000);

}

</script>

</body>
</html>
"""

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=8080)
```
